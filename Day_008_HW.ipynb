```python
data = {
    'AMT_INCOME_TOTAL': app_train['AMT_INCOME_TOTAL'],
}
dataset = pd.DataFrame(data)

dataset
```

```python
dataset.reset_index().plot.scatter(x = 'index', y = 'AMT_INCOME_TOTAL')
```

```python
#使用hist()可畫出直方圖,可看出有資料與其他差距很大故其他的不明顯
app_train['AMT_INCOME_TOTAL'].hist()
plt.xlabel('AMT_INCOME_TOTAL')
```

```python
#在loc做判斷式,把佔筆過重的用.quantile(0.99)濾掉
test =app_train.loc[app_train['AMT_INCOME_TOTAL'] < app_train['AMT_INCOME_TOTAL'].quantile(0.99)]
test['AMT_INCOME_TOTAL'].hist()
plt.xlabel('AMT_INCOME_TOTAL')
```
